```python
import pathlib
import collections
import matplotlib.pyplot as plt
import seaborn
import numpy
from PIL import Image
import random
```

```python
DATA_DIR = pathlib.Path.cwd().parent/"data"/"raw"/"Driver Drowsiness Dataset (DDD)"
DROWSY_PATH = DATA_DIR/"Drowsy"
NON_DROWSY_PATH = DATA_DIR/"Non Drowsy"
```

```python
drowsy_images = [image for image in DROWSY_PATH.iterdir()]
non_drowsy_images = [image for image in NON_DROWSY_PATH.iterdir()]
```

```python
print(f"Drowsy Images: {len(drowsy_images)}\nNon Drowsy Images: {len(non_drowsy_images)}\nTotal Images: {len(drowsy_images)+len(non_drowsy_images)}")
```

```python
seaborn.barplot(x=["Drowsy","Non Drowsy"],y=[len(drowsy_images),len(non_drowsy_images)])
plt.title("Data Distribution")
plt.ylabel("No of Images")
plt.show()
```

```python
SAMPLE = 20
rand_drowsy_imgs = random.sample(drowsy_images,SAMPLE)
rand_non_drowsy_imgs = random.sample(non_drowsy_images,SAMPLE)

for i in range(SAMPLE):
    if Image.open(rand_drowsy_imgs[i]).size == (227,227) and Image.open(rand_drowsy_imgs[i]).mode == "RGB":
        print("Test Passed")
    else:
        print(f"Test Failed for image: {rand_drowsy_imgs[i]}")
    if Image.open(rand_non_drowsy_imgs[i]).size == (227,227) and Image.open(rand_non_drowsy_imgs[i]).mode == "RGB":
        print("Test Passed")
    else:
        print(f"Test Failed for image: {rand_non_drowsy_imgs[i]}")
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 5, figsize = (10,5))
for i, axis in enumerate(axes):
    for idx, j in enumerate(axis):
        if i == 0:
            j.imshow(Image.open(rand_drowsy_imgs[idx]))
            j.set_title("Drowsy")
            j.axis("off")
        elif i == 1:
            j.imshow(Image.open(rand_non_drowsy_imgs[idx]))
            j.set_title("Non Drowsy")
            j.axis("off")
plt.tight_layout()
```
